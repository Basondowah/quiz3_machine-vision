```python
# Mentioning the required libraries
import matplotlib.pyplot as plt
import numpy as np
import cv2
```

```python
image = cv2.imread('eye.webp')

# To convert the image to grayscale:
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# To Binarize the image using Otsu's thresholding
threshold, binarized = cv2.threshold(gray, 0, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)
```

```python
# To convert the image to an edge map using Canny edge detection
edges = cv2.Canny(binarized, 100, 200)

# To preform the circular Hough transform
circles = cv2.HoughCircles(edges,cv2.HOUGH_GRADIENT,1,800,param1=30,param2=1,minRadius=0,maxRadius=0)
```

```python
if circles is not None:
    circles = np.round(circles[0, :]).astype("int")
    for (x, y, r) in circles:
        cv2.circle(image, (x, y), r, (0, 255, 0), 4)
```

```python
# To show the image
from google.colab.patches import cv2_imshow

im_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

cv2_imshow(im_rgb)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
